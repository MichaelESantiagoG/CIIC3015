```python
numem = int(input('Enter the number of employees: '))
salsum = 0
n = 1
while n <= numem:
    salary = float(input('Enter the salary of employee: $'))
    salsum = salsum + salary
    n = n + 1
average = salsum / numem
if average > 2500:
    print("You exceeded the limit for salaries.")
else:
    print("Average salary of the employees: $",average, sep="")

# Question author's solution (Python3):


num_employees = int(input('Enter the number of employees: '))
accumulator = 0
salary = 2500
counter = num_employees
while counter: 
    counter -= 1
    accumulator += float(input('Enter the salary of employee: $'))

average = accumulator / num_employees
if average > 2500:
  print('You exceeded the limit for salaries.')
else:
  print('Average salary of the employees: $' , average, sep='')
```
