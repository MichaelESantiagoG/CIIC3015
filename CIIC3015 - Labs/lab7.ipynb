```python
dlist =[]
lenoflist = int(input('Enter the length of the list: '))
i = 1
while i <= lenoflist:
  item = float(input('Enter the number that you will add: '))
  dlist.append(item)
  i += 1
dlist.sort()
print('Ordered list is:  ',dlist,sep='',end='')
```

```python
def string_test(string):
  if string.isalpha():
    print('Contains only alphabetic letters')
  if string.islower():
    print('It is written entirely in lowercase')
  if string.isupper():
    print('It is written entirely in uppercase')
  if string.isdigit():
    print('Contains only numeric digits')
  if string.startswith(' '):
    print('It starts with a space')   
  if string.endswith(' '):
    print('It ends with a space')

a = "fred"
string_test(a)

b = " sToRmaGeDdOn, DaRk LoRd Of AlL "
string_test(b)

c = "Maria"
string_test(c)

d = "CARMEN "
string_test(d)

e = "  PEdro    "
string_test(e)

f = "1234"
string_test(f)

g = "123abc"
string_test(g)

h = "a b c 1 2 3 "
string_test(h)

i = ""
string_test(i)

j = " "
string_test(j)
```

```python
def string_operations(string):
  sl =[]
  a = string.lstrip()
  sl.append(a)
  b = string.rstrip()
  sl.append(b)
  c = string.strip()
  sl.append(c)
  d = c.upper()
  sl.append(d)
  e = c.lower()
  sl.append(e)
  return sl

s = "fred"
result = string_operations(s)
for r in result: print("~"+r+"~")

s = " sToRmaGeDdOn, DaRk LoRd Of AlL "
result = string_operations(s)
for r in result: print("~"+r+"~")

	
s = "Maria"
result = string_operations(s)
for r in result: print("~"+r+"~")

	
s = "CARMEN "
result = string_operations(s)
for r in result: print("~"+r+"~")

	
s = "  PEdro    "
result = string_operations(s)
for r in result: print("~"+r+"~")

	
s = "1234"
result = string_operations(s)
for r in result: print("~"+r+"~")

	
s = "123abc"
result = string_operations(s)
for r in result: print("~"+r+"~")

	
s = "a b c 1 2 3 "
result = string_operations(s)
for r in result: print("~"+r+"~")

	
s = ""
result = string_operations(s)
for r in result: print("~"+r+"~")

	
s = ""
result = string_operations(s)
for r in result: print("~"+r+"~")

s = " "
result = string_operations(s)
for r in result: print("~"+r+"~")

```
