```python
def factorial(number):
  if number < 0:
    return -1
  elif number == 0:
    return 1
  else:
    i =  1
    accum = number
    while i < number:
      accum = accum * (number - i)
      i += 1
    return accum

# Question author's solution (Python3):

def factorial(n):
    if n < 0:
        return -1
    else:
        p = 1
        i = 2
        while i <= n:
            p *= i
            i += 1
        return p


print(factorial(5))
print(factorial(-5))
print(factorial(0))
```
